```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
df= pd.read_csv('AccidentesAviones.csv', index_col=0)
```

```python
df.head()
```

```python
df= df.drop(columns=['flight_no', 'route', 'registration', 'cn_ln', 'summary'])
```

```python
df.head(10)
```

```python
df.rename(columns={'fecha':'Fecha','HORA declarada': 'Hora Declarada', 'OperadOR': 'Operador','ac_type': 'Tipo de Avion', 'all_aboard': 'Total a Bordo','PASAJEROS A BORDO': 'Pasajeros a Bordo', 'crew_aboard': 'Tripulacion a Bordo', 'cantidad de fallecidos': 'Cantidad de Fallecidos','passenger_fatalities': 'Pasajeros Fallecidos', 'crew_fatalities': 'Tripulantes Fallecidos', 'ground':'Fatalidades en Tierra'}, inplace=True)
```

```python
df.head()
```

```python
df.replace("?",np.nan, inplace=True)
```

```python
df[['Mes','Dia','Año']]=df['Fecha'].str.split(' ', expand=True)
```

```python
df['Dia']=df.Dia.str.replace(',','')
```

```python
df['Mes']=pd.to_datetime(df['Mes'], format='%B').dt.month
```

```python
df['Fecha']=df['Año'].astype(str)+'-'+df['Mes'].astype(str)+'-'+df['Dia'].astype(str)
```

```python
df.head()
```

```python
df[df['Hora Declarada'].str.len()>4]
```

```python
df['Hora Declarada']=df['Hora Declarada'].str.replace(':','')
```

```python
df['Hora Declarada']=df['Hora Declarada'].str.replace('c','')
```

```python
df['Hora Declarada']=df['Hora Declarada'].str.strip()
```

```python
df['Hora Declarada']=df['Hora Declarada'].str.replace('Z','')
```

```python
df['Hora Declarada']=df['Hora Declarada'].str.replace(';','')
```

```python
df.head()
```

```python
df['Hora Declarada']=pd.to_datetime(df['Hora Declarada'], format='%H%M').dt.time
```

```python
df.head()
```

```python
df['Fecha']=pd.to_datetime(df['Fecha'],format='%Y-%m-%d',errors='coerce')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(subset='Total a Bordo',inplace=True)
```

```python
df=df.reset_index(drop=True)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.to_csv('Accidentes_Transformado.csv', index= False)
```
